```
#Nama : Eraraya Morenzo Muten
#NIM  :16520002
#Latihan 4 Pengulangan

#Program AnakAyam-1
#Menuliskan lagu anak ayam berdasarkan masukan jumlah anak ayam
#Versi menggunakan while

#KAMUS
#N = integer
#i = integer

#ALGORITMA
N = int(input("Masukan jumlah anak ayam = "))

print("Anak ayam turunlah " + str(N))
i = N-1                                   #perlu membuat variabel kondisi
while (i >= 1):                           #kriteria lebih ringkas
  print("Mati satu tinggallah " + str(i)) 
  i-=1                                    #perlu mengubah variabel kondisi
                                          #kalimat tidak selalu sama, sehingga
print("Mati satu tinggalah induknya")       #lebih tepat menggunakan while-loop
```

```
#Nama : Eraraya Morenzo Muten
#NIM  :16520002
#Latihan 5 Pengulangan

#Progam Hampiran Luas di Bawah Kurva
#Program ini menghitung luas di bawah kurva dengan pendekatan trapesium

#KAMUS
#a, b, d, x1, x2, t1, t2, ls  = float

#ALGORITMA
#assign variabel dengan input
a = float(input())
b = float(input())
d = float(input())
#assign variabel ls dengan nilai 0
ls = 0

#lakukan perhitungan luas trapesium

for i in range(int((b-a)//d)): #(b-a)//d adalah jumlah trapesium dengan lebar d 
                               #dengan asumsi (b-a) selalu habis dibagi d
  x1 = (a+d*i)                 #titik x pertama adalah a, lalu a+d, a+2d, dst
  t1 = x1**3 + x1 + 1          #sisi sejajar trapesium ke-1 adalah f(x1) 
  x2 = x1 + d                  #titik x kedua berjarak d dari x1
  t2 = x2**3 + x2 + 1          #sisi sejajar trapesium ke-2 adalah f(x2)
  ls = ls + (t1 + t2)*d/2      #menjumlahkan semua luas trapesium siku-siku
  
print(ls)
```
