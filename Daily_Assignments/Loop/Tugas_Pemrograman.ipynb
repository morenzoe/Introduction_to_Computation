```
#Nama : Eraraya Morenzo Muten
#NIM  :16520002
#Latihan 3 Pemrograman

#Progam Jumlah Bilangan Genap
#Program ini menghitung total bilangan genap antara 3 sampai dengan 30

#KAMUS
#i = integer
#total = integer

#ALGORITMA
total = 0                                                                       #inisialisasi
for i in range(3,31):                                                           #pengulangan dengan for antara angka dari 3 sampai dengan 30
  if i%2 == 0:                                                                  #jika i adalah bilangan genap
    total = total + i                                                           #aksi
print(total)                                                                    #terminasi
```
