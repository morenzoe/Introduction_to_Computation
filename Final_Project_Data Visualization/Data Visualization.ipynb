# Deskripsi Tugas 4: Karakteristik Data

Membuat DataFrame nama kolom dan jenis data

```
cdf.dtypes
```

```
list_nama_kolom = list(cdf.columns)

list_tipe_data = []

for i in cdf.dtypes:
  list_tipe_data.append(i)

dict_karakterisik = {'Nama Kolom' : list_nama_kolom, 'Tipe Data' : list_tipe_data}

df_karakteristik = pd.DataFrame(dict_karakterisik)
```

```
df_karakteristik
```

Tipe data numpy object termasuk dalam jenis data kategorikal. Pada dataset ini, jenis data kategorikal terdiri dari kategorikal nominal dan kategorikal ordinal.

Jenis data kategorikal nominal adalah data yang tidak dapat dinyatakan bahwa kategori yang satu lebih baik dari kategori lainnya atau dengan kata lain kategori yang tidak memiliki urutan tertentu. Berdasarkan nilainya, kolom yang termasuk dalam kategorikal nominal adalah Segment, Country, dan Product. 

Sedangkan jenis data kategorikal nominal memiliki urutan tertentu. Berdasarkan nilainya, kolom yang teramsuk dalam kategorikal ordinal adalah DiscountBand, Date, dan MonthName.

Selain itu, tipe data int64 dan float64 termasuk dalam jenis data kuantitatif.

```
list_jenis_data = []

for i in range(len(df_karakteristik)):
  if df_karakteristik['Nama Kolom'][i] in ['Segment', 'Country', 'Product']:
    list_jenis_data.append('Kategorikal Nominal')
  elif df_karakteristik['Nama Kolom'][i] in ['DiscountBand', 'Date', 'MonthName']:
    list_jenis_data.append('Kategorikal Ordinal')
  elif df_karakteristik['Tipe Data'][i] == 'float64' or df_karakteristik['Tipe Data'][i] == 'int64':
    list_jenis_data.append('Kuantitatif')
  
dict_jenis_data = {'Jenis Data' : list_jenis_data}

df_karakteristik = df_karakteristik.join(pd.DataFrame(dict_jenis_data))
df_karakteristik
```

##Karakteristik Jenis Data Kategorikal Nonimal

```
nilai_Segment = []
for i in cdf['Segment']:
  if i not in nilai_Segment:
    nilai_Segment.append(i)

nilai_Country = []
for i in cdf['Country']:
  if i not in nilai_Country:
    nilai_Country.append(i)

nilai_Product = []
for i in cdf['Product']:
  if i not in nilai_Product:
    nilai_Product.append(i)
```

```
print("Nilai-nilai yang ada di atribut Segment adalah:")
for i in nilai_Segment:
  print(i)
print()

print("Nilai-nilai yang ada di atribut Country adalah:")
for i in nilai_Country:
  print(i)
print()

print("Nilai-nilai yang ada di atribut Product adalah:")
for i in nilai_Product:
  print(i)
print()
```

##Karakteristik Jenis Data Kategorikal Ordinal

```
df_nilai_DiscountBand = pd.DataFrame(cdf['DiscountBand'].value_counts()).reset_index()
df_nilai_DiscountBand.rename(columns = {'index':'Nilai', 'DiscountBand' : 'Jumlah'}, inplace = True) 
df_nilai_DiscountBand
```

```
df_nilai_Date = pd.DataFrame(cdf['Date'].value_counts()).reset_index()
df_nilai_Date.rename(columns = {'index':'Nilai', 'Date' : 'Jumlah'}, inplace = True)
df_nilai_Date['Nilai'] = pd.to_datetime(df_nilai_Date['Nilai'], format='%d/%m/%Y')
df_nilai_Date.sort_values(by = 'Nilai', inplace = True)
df_nilai_Date
```

```
df_nilai_MonthName = pd.DataFrame(cdf['MonthName'].value_counts()).reset_index()
df_nilai_MonthName.rename(columns = {'index':'Nilai', 'MonthName' : 'Jumlah'}, inplace = True) 

sorter = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df_nilai_MonthName.Nilai = df_nilai_MonthName.Nilai.astype("category")
df_nilai_MonthName.Nilai.cat.set_categories(sorter, inplace = True)
df_nilai_MonthName.sort_values(by = 'Nilai', inplace = True)
df_nilai_MonthName
```
