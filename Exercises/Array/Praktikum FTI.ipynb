## Manual Append Problem

Screenshot_20201119-161256_Xodo Docs.jpg

```
N = int(input("Masukkan n: "))                                                  # deklarasi variabel panjang array, asumsi N integer tidak negatif

A = [0 for i in range(N)]                                                       # deklarasi array awal

for i in range(N):                                                              # loop buat ngisi array awal, asumsi elemen integer
  A[i] = int(input("Masukkan elemen ke " + str(i+1) + ": "))
  
NE = int(input("Masukkan elemen baru: "))                                       # deklarasi variabel elemen baru, asumsi NE integer

NEi = int(input("Masukkan posisi elemen baru: "))                               # deklarasi variabel indeks elemen baru

while NEi > N+1:                                                                # loop memastikan indeks elemen baru tidak melebihi panjang array baru (N+1)
  print("Indeks elemen di luar jangkauan. Ulangi!")
  NEi = int(input("Masukkan posisi elemen baru: "))

NA = [NE-1 for i in range(N+1)]                                                 # deklarasi array baru dengan panjang N+1 dan isi angka selain elemen baru

NA[NEi-1] = NE                                                                  # assign elemen indeks NEi di array baru dengan elemen baru

j = 0                                                                           # deklarasi variabel j untuk indeks akses array awal nanti
for i in range(N+1):                                                            # loop untuk memeriksa semua elemen array baru
  if NA[i] == NE-1:
    NA[i] = A[j]                                                                # jika elemen tersebut bukan elemen baru, isi dengan elemen array lama
    j+=1                                                                        # setelah terpakai, lanjut ke elemen array awal setelahnya

print(NA)                                                                       # cetak array baru 
```

## Triple Sum Problem

Screenshot_20201119-161306_Xodo Docs.jpg

```
N = int(input("Masukkan nilai N: "))                                            # deklarasi variabel panjang array, asumsi N integer tidak negatif

X = int(input("Masukkan nilai X: "))                                            # deklarasi variabel jumlah triple, asumsi X integer

A = [0 for i in range(N)]                                                       # deklarasi array 

for i in range(N):                                                              # loop buat ngisi array, asumsi elemen integer
  A[i] = int(input("Masukkan elemen ke " + str(i+1) + ": "))

if N<3:
  print("Ada 0 buah tripel.")                                                   # jika array tidak memiliki 3 elemen, pasti tidak ada tripel
else:
  S = 0                                                                         # deklarasi variabel S untuk total tripel
                                                                                # nested loop
  for i in range(0, N-2):                                                       # variabel i memeriksa dari angka pertama sampai 3 angka dari belakang
    for j in range(i+1, N-1):                                                   # variabel j memeriksa dari setelah i sampai 2 angka dari belakang
      for k in range(j+1, N):                                                   # variabel k memeriksa dari setelah j sampai angka terakhir
        if A[i]+A[j]+A[k]==X:                                                   # jika ada jumlah tripel yang sama dengan X, tambahkan 1 pada variabel S
          S+=1                                                                    

  print("Ada", S, "buah tripel.")                                               # cetak total tripel yang ada
```

## Sorting, Max, Min Problem

SmartSelect_20201119-161328_Xodo Docs.jpg

```
N = int(input("Masukkan n: "))                                                 # deklarasi variabel panjang array, asumsi N integer tidak negatif

A = [0 for i in range(N)]                                                       # deklarasi array awal

for i in range(N):                                                              # loop buat ngisi array awal, asumsi elemen integer
  A[i] = int(input("Masukkan elemen ke " + str(i+1) + ": "))
print(A)
Max = -100                                                                       # tabel frekuensi untuk semua angka dari terkecil hingga terbesar
Min = 100

for i in range(N):                                                             # memeriksa semua elemen array untuk mencari nilai maksimum dan minimum 
    if A[i] > Max:                                                              # jika anagram, nilai maksimum dan minimum array A sama dengan array B, cukup periksa salah satu saja
      Max = A[i]
    if A[i] < Min:
      Min = A[i]
print(Max, Min)
FA = [0 for i in range(Min, Max+1)]                                             # buat variabel array untuk menampung frekuensi seluruh angka dari Min sampai Max

for i in range(Min, Max+1):                                                     # periksa semua angka dari Min sampai Max
  for j in range(N):       
    if i == A[j]:                                                               # bandingkan dengan semua elemen array A
      FA[i-abs(Min)] +=1                                        # tambahkan 1 pada elemen frekuensi masing-masing untuk setiap elemen array A yang sama dengan angka uji
print(FA)
AS = [0 for i in range(N)]
print(Max+1-Min)
k=0
for i in range(Max+1-Min):
  for j in range(FA[i]):
    AS[k] = i+Min
    print(AS)
    k+=1
print(AS)

MM = [0 for i in range(N)]

j=0
for i in range(N):
  if i%2==0:
    MM[i]=AS[j]
    j+=1
  else:
    MM[i]=AS[-j]
print(MM)
```
