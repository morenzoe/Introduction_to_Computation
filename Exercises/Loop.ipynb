# Filling bucket z with bucket x and bucket y

1604418763172.jpg

```
x = int(input("Masukkan x: "))
y = int(input("Masukkan y: "))
z = int(input("Masukkan z: "))

i = 0

while (z-i*x)%y != 0 and z-i*x>0:
  i+=1                            

if (z-i*x)%y==0:
  print(i, "kali ember 1,", int((z-i*x)/y), "kali ember 2")

else:
  print("Tidak mungkin")
```
