Import Library

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Membaca File CSV

```
# membaca file csv pada directory yang sama dengan notebook ini sebagai DataFrame Pandas
df = pd.read_csv('nilai_uas.csv')
```

Nomor 1

```
# membuat grafik histogram dengan frekuensi per 10 nilai
hist = df['nilai_kal'].plot.hist(bins=range(0, 110, 10))
# menamai judul grafik
hist.set_title('Histogram Nilai Kalkulus')
# mengatur angka pada sumbu x
hist.set_xticks(np.arange(0,110,10))
# menamai sumbu x
hist.set_xlabel('Nilai')
# menamai sumbu y
hist.set_ylabel('Frekuensi')
# menampilkan histogram distribusi nilai kalkulus
plt.show()
```

Nomor 2

```
# membuat DataFrame nama fakultas dan jumlahnya
df_fakultas = df['fakultas'].value_counts()

# membuat diagram batang horizontal banyaknya mahasiswa masing-masing fakultas
barh = df_fakultas.plot(kind='barh', legend=None)
# menamai judul grafik
barh.set_title('Persebaran Mahasiswa')
# menamai sumbu x
barh.set_xlabel('Jumlah')
# menamai sumbu y
barh.set_ylabel('Fakultas')
# menampilkan grafik
plt.show()
```

03_02(3).png

Nomor 3

```
# membuat DataFrame nama fakultas dan jumlahnya
df_fakultas = df['fakultas'].value_counts()

# membuat diagram pie banyaknya mahasiswa masing-masing fakultas
pie = df_fakultas.plot(kind='pie', autopct='%1.1f%%')
# menamai judul grafik
pie.set_title('Persebaran Mahasiswa')
# menghilangkan nama sumbu y
pie.set_ylabel('Fakultas')
# mengatur posisi nama sumbu y
pie.yaxis.set_label_coords(-0.1,0.5)
# menampilkan grafik
plt.show()
```

03_03(1).png

Nomor 4

Berdasarkan diagram pie dan batang horizontal, fakultas dengan mahasiswa terbanyak adalah fakultas STEI.

Diagram batang horizontal lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak. Dengan 12 fakultas dan jumlah yang tidak terlalu jauh berbeda, diagram pie tidak merepresentasikan jumlah yang mudah dibandingkan, sedangkan diagram batang horizontal dengan sangat jelas menampilkan perbandingan urutan besar nilai masing-masing fakultas meskipun tanpa keterangan angka.

Nomor 5

```
# membuat scatter plot dengan nilai kimia sebagai x dan fisika sebagai y
# membuat titik data semi-transparan agar menghindari penumpukan dan densitas data tervisualisasikan
scatter = df.plot(kind='scatter', x='nilai_kim', y='nilai_fis', alpha=0.25)
# menamai judul grafik
scatter.set_title('Scatter Plot Nilai Kimia dan Nilai Fisika')
# menamai sumbu x
scatter.set_xlabel('Nilai Kimia')
# menamai sumbu y
scatter.set_ylabel('Nilai Fisika')
# menampilkan grafik
plt.show()
```
